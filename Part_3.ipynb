```python
import cv2
import time
import hand as htm
import math
from comtypes import CLSCTX_ALL
from pycaw.pycaw import AudioUtilities, IAudioEndpointVolume
import numpy as np

cap = cv2.VideoCapture(0)
pTime = 0
detector = htm.handDetector(detectionCon=1)

devices = AudioUtilities.GetSpeakers()
interface = devices.Activate(
    IAudioEndpointVolume._iid_, CLSCTX_ALL, None)
volume = interface.QueryInterface(IAudioEndpointVolume)
# volume.GetMute()
# volume.GetMasterVolumeLevel()
volRange =volume.GetVolumeRange()
minVol = volRange[0]
maxVol = volRange[1]
while True:
    ret, frame = cap.read()
    frame = detector.findHands(frame)
    frame = cv2.resize(frame, (0, 0), fx=1.1, fy=1.1)
    lmList = detector.findPosition(frame, draw=False)

    cTime = time.time()
    fps = 1/(cTime-pTime)
    pTime=cTime
    cv2.putText(frame, f"FPS: {int(fps)}", (50, 50), cv2.FONT_HERSHEY_PLAIN, 3, (0, 0, 255), 2)
    if len(lmList)!=0:
        x1 = lmList[4][1]
        y1 = lmList[4][2]
        x2 = lmList[8][1]
        y2 = lmList[8][2]
        img = cv2.line(frame, (x1, y1), (x2, y2), (0, 0, 255), 2)
        img = cv2.circle(frame, (x1, y1), 10, (0, 255, 0), -1)
        img = cv2.circle(frame, (x2, y2), 10, (0, 255, 0), -1)
        img = cv2.circle(frame, ((x1+x2)//2, (y1+y2)//2), 10, (0, 255, 0), -1)
        len_ = round(math.sqrt((x2-x1)**2+(y2-y1)**2))
        # Khoảng cách 2 ngón 25 đến 230
        # dải âm thanh từ -65 đến 0
        vol = np.interp(len_, [25, 230], [minVol, maxVol])
        volBar = np.interp(vol, [minVol, maxVol], [280, 100])
        print(volBar)
        volume.SetMasterVolumeLevel(vol, None)
        img = cv2.rectangle(frame, (50, int(volBar)), (80, 280), (0, 255, 0), -1)
        if len_ < 25:
            img = cv2.circle(frame, ((x1+x2)//2, (y1+y2)//2), 10, (255, 255, 0), -1)
    img = cv2.rectangle(frame, (50, 100), (80, 280), (0, 255, 0), 2)
    cv2.imshow("Cửa sổ hiển thị", frame)

    if cv2.waitKey(1) == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
